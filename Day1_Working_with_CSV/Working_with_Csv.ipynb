<h3>14.Converters</h3>

```python
pd.read_csv('ipl_match.csv')
```

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    elif name =="Mumbai Indians":
        return "MI"
    elif name =="Chennai Super Kings":
        return "CSK"
    elif name =="Rajasthan Royals":
        return "RR"
    elif name =="Kolkata Knight Riders":
        return "KKR"
    elif name =="Delhi Daredevils":
        return "DD"
    elif name =="Sunrisers Hyderabad":
        return "SRH"
    elif name =="Delhi Capitals":
        return "DC"
    elif name == "Deccan Chargers":
        return "DCC"
    else:
        return 
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv('ipl_match.csv',converters={'team1':rename,'team2':rename,'toss_winner':rename,'winner':rename}).head()
```
